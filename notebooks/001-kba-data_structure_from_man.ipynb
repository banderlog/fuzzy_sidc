```python
def int_to_str_values(a):
    # recursive function
    for k, v in a.items():
        if isinstance(v, dict):
            int_to_str_values(v)
        elif isinstance(v, int):
            a.update({k: str(v)})
```

```python
def process_write(ent, mod1, mod2, output):
    int_to_str_values(ent)
    int_to_str_values(mod1)
    int_to_str_values(mod2) 

    tmp = {
        "entity": ent,
        "modifier_1": mod1,
        "modifier_2": mod2
    }

    with open(output, 'w') as fp:
        json.dump(tmp, fp, ensure_ascii=False, indent=2)
```
